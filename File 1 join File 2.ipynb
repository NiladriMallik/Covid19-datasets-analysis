```python
import pandas as pd
import numpy as np
```

```python
output_path = 'OUTPUT DATAFRAME EXCEL SHEETS'
df_india_null_filled_path = output_path + '/df_india_null_filled.xlsx'
df_india_null_filled = pd.read_excel(df_india_null_filled_path)
df_india_null_filled
```

```python
covid_vaccine_statewise = 'covid_vaccine_statewise.csv'
df_covid_vaccine_statewise = pd.read_csv(covid_vaccine_statewise)
df_covid_vaccine_statewise
```

```python
df_cmbnd1_2_inner_temp = pd.merge(df_india_null_filled, df_covid_vaccine_statewise, 
                             how = 'inner', left_on = 'State', right_on = 'STATE',
                            suffixes = ('_left','_right'))
df_cmbnd1_2_inner_temp.head()
```

```python
df_cmbnd1_2_inner_temp.columns
```

```python
df_cmbnd1_2_inner_temp.drop('Unnamed: 0',axis = 1, inplace = True)
df_cmbnd1_2_inner_temp.head()
```

```python
df_cmbnd1_2_inner_temp.set_index('Sno', inplace = True)
df_cmbnd1_2_inner_temp.head()
```

```python
df_cmbnd1_2_inner = df_cmbnd1_2_inner_temp.drop('STATE',axis = 1, inplace = False)
df_cmbnd1_2_inner.head()
```
